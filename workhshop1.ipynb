```python
def fib(n):
    a, b = 0, 1
    for i in range(n):
        print(a)
        # a, b = b, a + b
        temp = a
        a = b
        b = temp + b
term = int(input("Enter your term: "))
fib(term)
```

```python
def myPow(x, n):
  result = 1
  i = 0
  while i < n:
    result *= x
    i += 1
  return result

print(myPow(4, 3))
```

```python
def sumDigits(n):
    total = 0
    for digit in str(n):
        total = total + int(digit)
    return total

print(sumDigits(234))
```

```python
def squaredSum(numbers):
    total = 0
    for n in numbers:
        total = total + (n * n)
    return total

print(squaredSum([2, 3, 4]))
```

```python
def sumDigits(num):
    total = 0
    for i in range(num + 1):
        total = total + i
    return total

print(sumDigits(6))
```
