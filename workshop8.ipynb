```python
import numpy as np

matrix_a = np.random.randint(0, 10, size=(2, 3))

matrix_b = np.random.randint(0, 10, size=(2, 3))

matrix_c = np.random.randint(0, 10, size=(2, 3))

result_matrix_new = (matrix_a - matrix_b) + matrix_c

print("Matrix A:")
print(matrix_a)

print("\nMatrix B:")
print(matrix_b)

print("\nMatrix C:")
print(matrix_c)

print("\nResult Matrix (A - B + C):")
print(result_matrix_new)
```

Task 3

```python
import numpy as np

matrix = np.random.randint(1, 21, size=(4, 4))
print("Generated 4x4 Matrix:")
print(matrix)

while True:
    try:
        user_number_str = input("\nEnter a number to check if it's in the matrix (1-20): ")
        user_number = int(user_number_str)
        if 1 <= user_number <= 20:
            break # Exit loop if input is valid
        else:
            print("Please enter a number between 1 and 20.")
    except ValueError:
        print("Invalid input. Please enter an integer.")

if user_number in matrix:
    count = np.count_nonzero(matrix == user_number)
    print(f"\nThe number {user_number} is in the matrix. It appears {count} time(s).")
else:
    print(f"\nThe number {user_number} is NOT in the matrix.")
```

```python
import numpy as np

A = np.random.randint(0, 10, size=(5, 5))
print("Matrix A:")
print(A)

rank_A = np.linalg.matrix_rank(A)
print(f"\nRank of Matrix A: {rank_A}")

if rank_A == 5:
    print("Matrix A is invertible because its rank is equal to its dimension (5).")
else:
    print("Matrix A is not invertible because its rank is less than its dimension (5).")
```

Task 5

```python
import numpy as np

A = np.random.randint(0, 10, size=(5, 5))
print("Matrix A:")
print(A)

x_true = np.random.randint(0, 10, size=(5, 1))
print("\nTrue Vector x:")
print(x_true)

b = np.dot(A, x_true)
print("\nVector b (A * x_true):")
print(b)

x_solved = np.linalg.solve(A, b)
print("\nSolved Vector x:")
print(x_solved)

print("\nVerification (x_true == x_solved):")
print(np.allclose(x_true, x_solved))
```

Task 6

```python
import numpy as np

A_system = np.array([
    [8, 3, -2],
    [-4, 7, 5],
    [3, 4, -12]
])

b_system = np.array([
    [9],
    [15],
    [35]
])

print("Coefficient Matrix A:")
print(A_system)
print("\nConstant Vector b:")
print(b_system)

solution = np.linalg.solve(A_system, b_system)

print("\nSolution (x, y, z):")
print(solution)

x_val, y_val, z_val = solution.flatten()
print(f"\nx = {x_val:.4f}")
print(f"y = {y_val:.4f}")
print(f"z = {z_val:.4f}")
```
