```python
import matplotlib.pyplot as plt
import math

# Step 1: Generate equidistant x-coordinates between the given interval
# Lists for coordinates

x = []
y = []


# Define start, end, and step size for x increment

start = -4 * math.pi

end = 4 * math.pi

step = 0.1 # Controls smoothness of the graph


i = start

while i <= end:

  x.append(i) # Store x-coordinate

# Step 2: Calculate the y-coordinate using sin function for each x

  y.append(math.sin(i))

  i += step


# Step 3: Plot the graph using calculated coordinates

# Label the graph properly

plt.plot(x, y, color="red", label="sin function")

plt.xlabel("X")

plt.ylabel("Y")

plt.title("y = sin(x) ,-4π<= x <= 4π")

plt.grid(True)

plt.legend()

plt.show()
```

```python
x =[] # List to store x values

y= [] # List to store y values

r = 4

center = [2,2]

# paramater theta start value

theta_start = 0.0

# parameter Theta end value (bit more than 2pi)

theta_end = 2.5*math.pi

# step sizes to sample the paramater Theta

step_size= 0.125

#math.ceil() → round up to nearest integer

# Number of samples

steps = math.ceil((theta_end-theta_start)/step_size)

# initialize the theta

# variable to increment Theta by each step

theta = theta_start

# for loop to calculate x and y value at each steps

for i in range(0,steps):

  x.append(r*math.cos(theta)+center[0])

  y.append(r*math.sin(theta)+center[1])

  theta += step_size

# create figure in new window

plt.figure()

# plot a dot a center

plt.plot(center[0],center[1],'.')

plt.plot(x,y,'-')

plt.grid()

plt.xlabel("X")

plt.ylabel("Y")

#plt.gca() → Get the current figure (axes object) for handling

ax =plt.gca()

#ensures 1 unit in x = 1 unit in y

#Without this ,figure is stretched(ircle could appear as an ellipse)

ax.set_aspect('equal',adjustable='box')

plt.show()
```

# 9. Plot the trigonometric COSINE function in interval [−3𝜋,3𝜋], i.e., 𝑦 = cos (𝑥), −3𝜋 ≤ 𝑥 ≤3𝜋

```python
import matplotlib.pyplot as plt
import math

# Generate x values from -3π to 3π
# smaller step for smoother curve

x = []
y = []

# Define start, end, and step size for x increment
start = -3*math.pi
end = 3*math.pi
step = 0.1

while start <= end:
    x.append(start)
    #calculating y-coordinate using cos function
    y.append(math.cos(start))
    start += step

# Plot the cosine function
plt.plot(x, y, color="Brown", label="y = cos(x)")

# Add labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Cosine Function in Interval [-3π, 3π]")

# Add grid and legend
plt.grid(True)
plt.legend()

plt.show()
```

# Circle

# 10. Plot a circle with center at (a, b) and radius r. The values of a=b=2, and r=4, using a while loop.

```python
import matplotlib.pyplot as plt
import math

# Lists to store coordinates
x = []
y = []

#radius
r = 4
#center
center = [2,2]

#assigning theta starting value
theta_start = 0.0
theta_end = 2.5*math.pi
# smaller step = smoother circle
step = 0.1
steps = math.ceil((theta_end-theta_start)/step)

# Generate points using parametric equations of a circle
# x = h + r*cos(θ), y = k + r*sin(θ)

theta = theta_start
while theta <= theta_end:
    x.append(center[0] + r*math.cos(theta))
    y.append(center[1] + r*math.sin(theta))
    theta += step

# Plot the circle
plt.plot(x, y, color="blue")

# Mark the center
plt.scatter(center[0], center[1], color="red", marker="x", label="Center")

plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Circle")
plt.legend()
plt.grid(True)
plt.show()
```

# Ellipse

# 11. Plot the above circle and the ellipse on the same axes. This is given an idea about the intersection points of the two conic sections.

```python
import matplotlib.pyplot as plt
import math

# Lists to store x and y values
x = []
y = []


a = 3   # semi-minor axis
b = 5   # semi-major axis
center = [2, 2]  # center (h, k)

theta_start = 0.0
theta_end = 3*math.pi
step_size = 0.125

# Number of samples
steps = math.ceil((theta_end - theta_start) / step_size)

# initializing theta
theta = theta_start

# calculating x and y coordinates
for i in range(steps):
    x.append(a*math.cos(theta) + center[0])  # x = a*cos(θ) + h
    y.append(b*math.sin(theta) + center[1])  # y = b*sin(θ) + k
    theta += step_size  # increment theta

# Plotting
plt.figure()

# plot the center point
plt.plot(center[0], center[1], '.', color="red")

# Plot the ellipse
plt.plot(x, y, '-', color="blue")

plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ellipse centered at (2,2)")

# Equal aspect ratio to avoid stretching
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Circle parameters
r = 4
circle_center = [2, 2]

# Lists for circle coordinates
x_circle = []
y_circle = []

theta = 0.0
theta_end = 3*math.pi
step = 0.1

while theta <= theta_end:
    x_circle.append(circle_center[0] + r*math.cos(theta))
    y_circle.append(circle_center[1] + r*math.sin(theta))
    theta += step

# Ellipse parameters
a = 3   # semi-minor axis
b = 5   # semi-major axis
ellipse_center = [2, 2]

x_ellipse = []
y_ellipse = []

theta = 0.0
theta_end = 3*math.pi
step = 0.1

while theta <= theta_end:
    x_ellipse.append(ellipse_center[0] + a*math.cos(theta))
    y_ellipse.append(ellipse_center[1] + b*math.sin(theta))
    theta += step

#Plotting both shapes together ---
plt.figure()

# Circle
plt.plot(x_circle, y_circle, color="blue", label="Circle")

# Ellipse
plt.plot(x_ellipse, y_ellipse, color="green" , label="Ellipse")

# Mark the common center
plt.scatter(2, 2, color="red", marker="x", label="Center (2,2)")
plt.legend(loc="upper right")

plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Circle and Ellipse on Same Axes")
plt.grid(True)
plt.show()
```

# Parabola

# 12. Plot a Parabola (𝑦 − 𝑝) = 4𝑎(𝑥 − 𝑞)^2 with vertex at (𝑝,𝑞). The values of p, q, and a can be chosen appropriately.

```python
import matplotlib.pyplot as plt
import math

# Parabola parameters
p = 2   # y-coordinate of vertex
q = 3   # x-coordinate of vertex
a = 0.5 # controls width and direction

# Step size for x values
step = 0.1

# Define x range around vertex
x_start = q - 5
x_end   = q + 5

# Lists to store coordinates
x_store = []
y_store = []

# Initialize x
x = x_start

# Generating points for parabola
while x <= x_end:
    y = p + 4*a*(x - q)**2   # equation: (y - p) = 4a(x - q)^2
    x_store.append(x)
    y_store.append(y)
    x += step

# Plot the graph
plt.figure(figsize=(6,6), dpi=100)
plt.title(f"Parabola with vertex at ({q},{p}) and a={a}")
plt.plot(x_store, y_store, '-', color="blue")
plt.scatter(q, p, color="red", marker="x", label="Vertex")
plt.grid(True)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc="center")
plt.show()
```
