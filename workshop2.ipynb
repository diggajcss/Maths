```python
def get_unique(numbers):
    unique = []
    for x in numbers:
        if x not in unique:
            unique.append(x)
    return unique

print(get_unique([1,2,3,3,3,3,4,5]))
```

```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
print(is_prime(11))
```

```python
for i in range(3, 999, 2):
    if is_prime(i) and is_prime(i + 2):
        print(i, "and", i + 2)
```

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total
print(sumOfDivisors(10))
```

```python
# Starts checking after 6
n = 7
while True:
    if sumOfDivisors(n) == n:
        print("Next perfect number is:", n)
        break
    n += 1
```

```python
marks = [20, 30, 35, 55, 70, 80, 90]

mean = sum(marks) / len(marks)

marks.sort()
n = len(marks)
if n % 2 == 0:
    median = (marks[n//2 - 1] + marks[n//2]) / 2
else:
    median = marks[n//2]

print("Mean:", mean)
print("Median:", median)
```

```python
def decimal_to_binary(n):
    if n == 0:
        return "0"

    binary = ""
    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2
    return binary

# Example
number = 25
print(f"Decimal: {number}")
print(f"Binary: {decimal_to_binary(number)}")
```

```python
def prodDigits(n):
    n = abs(n)
    product = 1
    for digit in str(n):
        product *= int(digit)
    return product

num = 143
print("Product of digits:", prodDigits(num))
```

```python
def multiply_all(*args):
    result = 1
    for num in args:
        result *= num
    return result

# Running the function for the requested sets:
set_a = multiply_all(1, 2, 3, 4, 5)
set_b = multiply_all(12, 13, 14)

print(f"Result for set (a): {set_a}")
print(f"Result for set (b): {set_b}")
```
