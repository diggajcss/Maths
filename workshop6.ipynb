Solve the linear system graphically using Python

```python


import numpy as np
import matplotlib.pyplot as plt

# x values
x = np.linspace(-5, 5, 200)

# Equations:
# 2x + 3y = 3   => y = (3 - 2x)/3
# 4x + 6y = 6   => y = (6 - 4x)/6
# 3x - 2y = 1   => y = (3x - 1)/2

y1 = (3 - 2*x) / 3
y2 = (6 - 4*x) / 6
y3 = (3*x - 1) / 2

# Plot lines
plt.plot(x, y1, label="2x + 3y = 3")
plt.plot(x, y2, label="4x + 6y = 6")
plt.plot(x, y3, label="3x - 2y = 1")

# Intersection of first two lines
# Since second equation is multiple of first,
# both overlap and intersect with third at (1, 1/3)

plt.scatter(1, 1/3, color="red")
plt.text(1, 1/3, " (1, 1/3)")

# Axes
plt.axhline(0, color="black")
plt.axvline(0, color="black")

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid(True)
plt.legend()
plt.title("Graphical Solution of Linear System")
plt.xlabel("x")
plt.ylabel("y")

plt.show()

print("Solution Point = (1, 1/3)")
```

```python
import numpy as np

M = np.random.randint(1, 21, (4, 4))

print("Random Matrix:")
print(M)

num = input("Enter a number: ")

if num.isdigit():
    num = int(num)
    count = np.count_nonzero(M == num)

    if count > 0:
        print(f"{num} found {count} time(s).")
    else:
        print(f"{num} not found.")
else:
    print("Invalid input!")
```

```python
import numpy as np

A = np.random.randint(1, 10, (5, 5))

print("Matrix A:")
print(A)

rank = np.linalg.matrix_rank(A)
print("Rank of Matrix:", rank)

if rank == 5:
    print("Matrix is invertible")
else:
    print("Matrix is not invertible")
```

```python
import numpy as np

A = np.random.randint(1, 10, (5, 5))
x = np.random.randint(1, 10, (5, 1))

print("Matrix A:")
print(A)

print("Vector x:")
print(x)

b = np.dot(A, x)

print("Vector b = Ax:")
print(b)

solution = np.linalg.solve(A, b)

print("Solved x:")
print(solution)
```
