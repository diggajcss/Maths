#Following data shows the change in population of different animals from 2017 to 2022.
# a. Create a pie chart for the population of each year
# b. Can we show all the pie chart on the single pane? Write the code.

```python
import matplotlib.pyplot as plt

years = [2017, 2018, 2019, 2020, 2021, 2022]
bears = [8, 54, 93, 116, 137, 184]
dolphins = [150, 77, 32, 11, 6, 1]
whales = [80, 54, 100, 76, 93, 72]

# Create 2x3 subplot grid
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
# Flatten to 1D array for easy indexing
axes = axes.flatten()

for i in range(len(years)):
    data = [bears[i], dolphins[i], whales[i]]
    labels = ['Bears', 'Dolphins', 'Whales']
    axes[i].pie(data, labels=labels, autopct='%1.1f%%')
    axes[i].set_title(f'{years[i]}')

plt.suptitle('Animal Population Distribution (2017–2022)', fontsize=18)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
bears     = [8, 54, 93, 116, 137, 184]
dolphins  = [150, 77, 32, 11, 6, 1]
whales    = [80, 54, 100, 76, 93, 72]
years     = [2017, 2018, 2019, 2020, 2021, 2022]

# 2017
plt.figure(figsize=(5,5))
plt.pie([bears[0], dolphins[0], whales[0]], labels=['Bears','Dolphins','Whales'], autopct='%1.1f%%')
plt.title('Animal Population in 2017')
plt.show()

# 2018
plt.figure(figsize=(5,5))
plt.pie([bears[1], dolphins[1], whales[1]], labels=['Bears','Dolphins','Whales'], autopct='%1.1f%%')
plt.title('Animal Population in 2018')
plt.show()

# 2019
plt.figure(figsize=(5,5))
plt.pie([bears[2], dolphins[2], whales[2]], labels=['Bears','Dolphins','Whales'], autopct='%1.1f%%')
plt.title('Animal Population in 2019')
plt.show()

# 2020
plt.figure(figsize=(5,5))
plt.pie([bears[3], dolphins[3], whales[3]], labels=['Bears','Dolphins','Whales'], autopct='%1.1f%%')
plt.title('Animal Population in 2020')
plt.show()

# 2021
plt.figure(figsize=(5,5))
plt.pie([bears[4], dolphins[4], whales[4]], labels=['Bears','Dolphins','Whales'], autopct='%1.1f%%')
plt.title('Animal Population in 2021')
plt.show()

# 2022
plt.figure(figsize=(5,5))
plt.pie([bears[5], dolphins[5], whales[5]], labels=['Bears','Dolphins','Whales'], autopct='%1.1f%%')
plt.title('Animal Population in 2022')
plt.show()
```

Construct a pie chart to visually display the favorite fruits of the students in a class based on the
given data: Mango - 45; Orange - 30; Plum - 15; Pineapple - 30; Melon – 30x

```python
import matplotlib.pyplot as plt

# Data
fruits = ['Mango', 'Orange', 'Plum', 'Pineapple', 'Melon']
counts = [45, 30, 15, 30, 30]

# Create pie chart
plt.figure(figsize=(7,7))
plt.pie(counts, labels=fruits, autopct='%1.1f%%', startangle=90)

plt.title('Favorite Fruits of Students')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
colors = ['Red', 'Silver', 'Blue', 'White', 'Green']
counts = [123, 230, 78, 193, 12]

# Bar chart
plt.figure(figsize=(8,6))
plt.pie(counts, labels=colors, autopct='%1.1f%%', startangle=0)
plt.title('Number of Cars by Color')
plt.xlabel('Car Color')
plt.ylabel('Number Seen')
plt.show()

```

```python
myexplode = [0.3, 0, 0, 0, 0]
plt.figure(figsize=(7,7))
plt.pie(counts, labels=colors, autopct='%1.1f%%', explode=myexplode, shadow=True)
plt.title('Car Color Distribution (Exploded Red with Shadow)')
plt.show()
```

```python

plt.figure(figsize=(7,7))
plt.pie(counts, autopct='%1.1f%%')
plt.legend(colors, title="Car Colours:")
plt.title('Car Color Distribution with Legend')
plt.show()


```

```python
myexplode = [0.3, 0, 0, 0, 0]
plt.figure(figsize=(7,7))
plt.pie(counts, labels=colors, autopct='%1.1f%%', explode=myexplode)
plt.title('Car Color Distribution (Exploded Red)')
plt.show()
```

```python
mycolours = ["magenta", "green", "cyan", "red", "yellow"]

plt.figure(figsize=(7,7))
plt.pie(counts, labels=colors, autopct='%1.1f%%', colors=mycolours)
plt.title('Car Color Distribution (Custom Colors)')
plt.show()
```
